c) Was sind Ausreißer? Wie können sie entstehen?

In der Statistik spricht man von einem Ausreißer, wenn ein Messwert nicht in eine erwartete Messreihe passt oder allgemein nicht den Erwartungen entspricht (Wikipedia). Ausreißer können durch einen Verfahrenstechnischen Fehler verursacht werden. Beispielsweise könnten die Daten falsch eingegeben sein oder es gab Messungenauigkeiten.  Darüber hinaus kann es sich um einen Wert handeln, der einen selten auftretende Messung richtig wiederspiegelt.

d) Welche der Statistiken in a) und b) sind robust bzw. nicht robust gegen Ausreißer? Begründen Sie. 

Der Median, Quantile und der Modus sind relativ robust gegen Ausreißer, weil sie ihren Wert bei wenigen Ausreißern nicht verändern. Der Durchschnitt und die Spannweite hingegen sind sehr empfindlich gegenüber Ausreißern. In den Durchschnitt fließen alle Werte gleichermaßen hinein. Gerade Ausreißer mit extremen Werten verändern den Durschnitt stark. Die Spannweite errechnet sich durch das Maximum und Minimum. In diesen Gebieten sind gerade die Ausreißer zu finden.

e) Erklären Sie, inwiefern Ausreißer die Analyse beeinflussen können.

Da Aureißer einige Statistiken stark beeinflussen, erschweren sie die Analyse von Datensätzen mittels solcher Metriken. Die Statistiken allein sind nicht ausreichend für die Exploration der Daten. Stattdessen muss man Visualisierungstechniken verwenden und die Datensätze von Ausreißern ggf. bereinigen. In einigen Fällen sind es jedoch gerade die Ausreißer, die es wert sind analysiert zu werden (wenn es sich nicht um Messfehler handelt).

f) Wie  kann  man  Ausreißer  in  Daten  erkennen?  Gibt es Ausreißer  im  Bodyfat-Datensatz? Visualisieren Sie sie,wenn möglich.

Man kann Ausreißer unter anderem mit statistischen Tests identifizieren. Diese setzen ggf. eine bestimmte Verteilung voraus (Dean-Dixon-Test, Normalverteilung; Nalimov-Test, Normalverteilung) und sie sind für eine unterschiedliche Anzahl von Messwerten zu gebrauchen.
Außerdem kann man Ausreißer anhand von Visualisierungen erkennen, zum Beispiel mithilfe von Boxplotdiagrammen. Oben erkennt man bei der Höhe, dass es einen starken Ausreißer nach unten gibt (29.5). Bei der Hüfte sieht man drei Ausreißer nach oben.
Ausreißer können aber auch multivariater Natur sein, wenn jeder Variablenwert für sich gesehen nicht ungewöhnlich ist, aber das Vorkommen der Variablenwerte zusammen ungewöhnlich sind. In der Scatter-Matrix unten sieht man bei der Dichte und Körperfett eine sehr starke Abhängigkeit voneinander. Doch über und unter der blauen Wolke aus Messwerten kann man einzelne Ausreißer als Punkte erkennen.

```python
from pandas.plotting import scatter_matrix
dummy = scatter_matrix(data, alpha=0.2, figsize=(20, 20), diagonal='kde')
```

g) Wie können wir mit Ausreißern im Allgemeinen umgehen? Wie werden Sie damit im Fall des Bodyfat-Datensatzes umgehen?

https://en.wikipedia.org/wiki/Outlier#Working_with_outliers
Anwendungen, die mit vielen Daten umgehen müssen, sollten robust gegenüber Ausreißern sein und bei diesen kann man die Ausreißer einfach im Datensatz behalten. Dies gilt vor allem dann, wenn keine Aussage über die tatsächliche Verteilung der Daten getroffen werden kann.
Sollen Ausreißer nicht einfach übernommen werden, können sie gelöscht werden oder die unnatürlichen Werte geeignet korrigiert werden. In jedem Fall sollten Ausreißer zunächst daraufhin untersucht werden, ob sie in der Domäne plausibel sind. Wird der Bodyfat-Datensatz im Laufe der Praktikums weiter verwendet, wird genau das die erste Aufgabe zur Untersuchung sein.
